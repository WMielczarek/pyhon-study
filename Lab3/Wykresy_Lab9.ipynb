```python
import csv
import math
import os
import numpy as np
import datetime
 
dates = []
times = []
id_station = []
datetimes = []
ztd_tab = []
x=[]
y=[]
 
for filename in os.listdir("dane"):
    file = ""
   # print(filename)
    file = "dane\\" + filename
    #print(file)
    f = open(file)
    csv_f = csv.reader(f)
 
    for row in csv_f:
        if row[0] == "WROC":
            date = row[1]
            time = row[2]
 
            datetime = row[1] + " " + row[2]
            id_statio = row[0]
            ztd = row[3]
  
            datetimes.append(datetime)
            ztd_tab.append(ztd)
            dates.append(date)
            times.append(time)
            id_station.append(id_statio)
        
     


datetimes = set(datetimes)
datetimes = sorted(datetimes)



parts = []
print(len(datetimes))


for part in range(0,24):
    parts.append(ztd_tab[part])
    print(part)
    print(parts[part])
    
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plt.axis('auto')
plt.xticks(rotation = 90)
plt.xlabel('Datetime')
plt.ylabel('ZTD Value')
plt.title('ZTD Chart from time')
print(len(datetimes))
plt.grid(True)
plt.plot(datetimes, parts,'ro')
#plt.plot(datetimes,temps)
plt.show()
f.close()
```
